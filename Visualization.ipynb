```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\Kaggle Data\melt\diabetic_data.csv") #importing dataset
```

```python
#Line graph
sns.lineplot(data=df, x="age", y='time_in_hospital').set(title='Age vs Time_in_hospital') #using seaborn to plot a lineplot
plt.plot() #plotting the graph
```

From this chart, we can infer that as the age increases, time in the hospital also increases. 

```python
#distribution plot 
sns.displot(data = df, x='num_lab_procedures', hue='number_diagnoses', kind='kde', fill=True).set(title='num_lab_procedures vs nuber_diagnoses') #using seaborn to plot distribution plot
plt.show()
```

From this plot, we can see that lab procedures increase as the number of diagnosis increase. 

```python
#boxplot graph 
sns.boxplot(data=df, x='number_diagnoses', y='num_medications').set(title='number_diagnoses vs num_medications') #plotting box plot
plt.show()
```

We can see that as the number of diagnoses increase, which means that diabetes has affected different regions. Hence the number of medications also increase to treat the patients. 

```python
#bar plot 
df = df[(df != '?').all(axis=1)] #cleaning the data by removing '?' from axis
sns.barplot(data=df, x='race', y='time_in_hospital', hue='gender').set(title='race vs time_in_hospital') #plotting barplot using seaborn
```

In this bar graph, we can see that race also plays a critical part in diabetes. Females get more diabetes as compared to males and AfricanAmerican people are more susceptable to get diabetes.
